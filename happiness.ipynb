```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
happiness_path = "data_files/WHR2021.csv"
df_happiness = pd.read_csv(happiness_path)
```

```python
df_happiness.head()
```

```python
df_happiness.shape
#(1949, 11)-11 attributes, 1949 samples
```

```python
df_happiness.info()
```

```python
current_year = df_happiness[df_happiness['year']==2020][:166]
plt.rcParams['figure.figsize']=(30, 7)
graph = sns.barplot(x = current_year['Country name'][:166], y = current_year['Life Ladder'][:166])
graph.set_xlabel(xlabel = 'Country', fontsize = 10)
graph.set_ylabel(ylabel = 'Happiness Index', fontsize = 10)
graph.set_title(label = 'Happiness by Country', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
